```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(42) # For reproducibility
X = np.random.rand(50, 1) * 10 # 50 samples, single feature
y = 3 * X + 5 + np.random.randn(50, 1) * 2 # Linear relationship with noise
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
print(f"Intercept (b): {model.intercept_[0]:.2f}")
print(f"Coefficient (w): {model.coef_[0][0]:.2f}")
```

**Comenting on the Intercept and Coefficient values**

5.19 and 2.96 for the intercept and coefficient, respectively, are the outcomes of the model with noise on the data.

```python
print(f"Regression Equation: y = {model.coef_[0][0]:.2f} * X + {model.intercept_[0]:.2f}")
```

```python
y_pred = model.predict(X)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color="blue", label="Data Points")
plt.plot(X, y_pred, color="red", label="Regression Line")
plt.title("Simple Linear Regression")
plt.xlabel("Feature (X)")
plt.ylabel("Target (y)")
plt.legend()
plt.grid(True)
plt.show()
```

**Commenting on the graph output**

The graph shows how good the linear regression model fits the data. The data points are a scatter of points, and the red regression line shows the best linear fit to these points.

The red line is computed by calculing a minimal value between the sum of the squared differences between the real values and the values predicted by the model.
